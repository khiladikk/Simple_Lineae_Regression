```python
#Importing the libraries
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Importing the dataset
data= pd.read_csv("Income_Data.csv")
```

```python
#Splitting into features and labels
features= data.iloc[:,:-1].values
```

```python
labels= data.iloc[:,-1].values
```

```python
#Splitting the dataset intp Training set and Test set
from sklearn.model_selection import train_test_split
```

```python
features_train,features_test, labels_train, labels_test= train_test_split(features, labels, test_size=0.2)
```

```python
#Fitting Simple Linear Regression to The Training set
from sklearn.linear_model import LinearRegression
regressor= LinearRegression()
regressor.fit(features_train, labels_train)
```

```python
#Predicting the Test set Results
labels_pred= regressor.predict(features_test)
```

```python
#Model Score
Score= regressor.score(features_test, labels_test)
```

```python
#Visualising the Training set Results
plt.scatter(features_train, labels_train, color= 'Blue')
plt.plot(features_train, regressor.predict(features_train), color = 'black')
plt.title('Income vs ML Experience(Training set)')
plt.xlabel('ML-Experience')
plt.ylabel('Income')
plt.show()
```

```python
#Visualising the Test set result
plt.scatter(features_test, labels_test, color="black")
plt.plot(features_test, regressor.predict(features_test), color= 'blue')
plt.title('Income vs ML Experience(Test Set)')
plt.xlabel("ML-Experience")
plt.ylabel('Income')
plt.show()
```
